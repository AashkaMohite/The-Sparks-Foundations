#### Ship Mode

```python
# value_count
df["Ship Mode"].value_counts()
```

```python
plt.figure(figsize = (6,6))
textprops = {"fontsize":15}
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.title('Ship Mode')
plt.pie(df['Ship Mode'].value_counts(), labels=df['Ship Mode'].value_counts().index,autopct='%1.1f%%',textprops = textprops, shadow=True, colors=colors)
plt.show()

```

```python
sns.countplot('Ship Mode',data=df,hue='Category')
```

```python
pls=df.groupby('Ship Mode')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
pls[:].plot.bar(colormap='viridis',figsize=(20,10))
plt.xlabel('Ship Mode')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```
